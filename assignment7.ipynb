```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Randomized Sales Data Generaon (NumPy)
np.random.seed(102317148)
sales_data = np.random.randint(1000, 5001, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)

#Data Manipulation and Analysis (Pandas)
print("the first 5 rows:",df.head())
print("summary statistics of the DataFrame",df.describe())
print("the total sales per category ",df.sum())
print("total sales per month.",df.sum(axis=1))
print("average sales growth between consecutive months:",df.pct_change().mean())
df['Growth Rate'] = df.sum(axis=1).pct_change() * 100
df['Electronics'] = df['Electronics'] * 0.9
print(df)

# Visualizations (Matplotlib and Seaborn)
plt.figure(figsize=(12, 6))
for category in categories:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.xlabel('Months')
plt.ylabel('Sales Units')
plt.title('Monthly Sales Trends')
plt.legend()
plt.grid()
plt.show()

# Box Plot of Sales Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.show()
```

```python
# floor, ceiling and truncated value, rounded values
arr_q3 = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_vals = np.floor(arr_q3)
ceil_vals = np.ceil(arr_q3)
trunc_vals = np.trunc(arr_q3)
rounded_vals = np.round(arr_q3)
print(floor_vals)
print(ceil_vals)
print(trunc_vals)
print(rounded_vals)
```

```python
#  swap two elements in a list in Python Using a Temporary Variable.
def swap_list_elements(lst, pos1, pos2):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp
    return lst
example_list = [1, 3, 8, 4, 5]
print("Before Swap:", example_list)
print("After Swap:", swap_list_elements(example_list, 1, 4))
```

```python
# swap two elements in a set by converng it to a list first.
def swap_set_elements(s, el1, el2):
    lst = list(s)
    i, j = lst.index(el1), lst.index(el2)
    lst[i], lst[j] = lst[j], lst[i]
    return set(lst)
ex_set={1, 3, 8, 4, 5}
swap_set_elements(ex_set, 8, 4)
print("Swapped Set:", ex_set)
```
